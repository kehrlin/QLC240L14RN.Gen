```python

# Create a figure and an axes with a specific projection
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Draw coastlines and countries for context
ax.coastlines()
ax.add_feature(cartopy.feature.BORDERS)

# Define the location for the arrow
lon, lat = -100, 30

# Draw the arrow
ax.annotate(
    '',
    xy=(-80.8431, 35.2271), xycoords='data',
    xytext=(lon + .05, lat), textcoords='data',
    arrowprops=dict(facecolor='black', shrink=0.05)
)
ax.text(lon, lat, 'We are here', transform=ax.transData, fontsize=10, color='black')

# Show the map
plt.show()
```
